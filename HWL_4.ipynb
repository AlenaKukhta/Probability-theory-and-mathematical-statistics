#### 1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
Найдите ее среднее значение и дисперсию.

**Математическое ожидание $M(X)$** равномерно распределенной непрерывной случайной величины можно вычислить по формуле:
$$M(X) = \frac{a + b}{2}$$

```python
M = (800 + 200) / 2

print('Среднее значение равномерно распределенной непрерывной случайной величины А = {:.4}'.format(M))
```

**Дисперсию $D(X)$** равномерно распределенной непрерывной случайной величины можно вычислить по формуле:
$$D(X) = \frac{(b - a)^2}{12}$$

```python
D = (800 - 200)**2 / 12

print('Дисперсия равномерно распределенной непрерывной случайной величины А = {:.6}'.format(D))
```

#### 2. О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
Если да, найдите ее.

По формуле **дисперсии $D(X)$** равномерно распределенной непрерывной случайной величины можно вычислить правую границу величины В:
$$D(X) = \frac{(b - a)^2}{12}$$
Следовательно:
$$b = a + \sqrt{D(X) * 12}$$

```python
D = 0.2
a = 0.5

b = a + math.sqrt(D * 12)

print('Правая граница величины В = {:.6}'.format(b))
```

**Математическое ожидание** можно вычислить по формуле:
$$M(X) = \frac{a + b}{2}$$

```python
M = (a + b) / 2

print('Среднее значение величины В = {:.6}'.format(M))
```

#### 3. Непрерывная случайная величина X распределена нормально и задана плотностью распределения

$$f(x) = \frac{1}{4 \sqrt{2 \pi}} e ^ {- \frac{(x + 2)^2}{32}},$$

или f(x) = (1 / (4 * sqrt(2*pi))) * (exp(-((x+2)**2) / 32))


Найдите:

а) M(X)

б) D(X)

в) std(X) (среднее квадратичное отклонение)

**Плотность вероятности нормального распределения** можно вычислить по формуле:
$$f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e ^ {- \frac{(x - a)^2}{2 \sigma^2}},$$
где $a = M(X), \; \sigma^2 = D(X).$


**Среднее квадратичное оклонение** можно вычислить по формуле:
$$\sigma = \sqrt{\frac{\sum\limits_{i=1}^{n} (x_i - \overline{x})^2}{n}}$$

Таким образом,

а) M(X) = - 2

б) D(X) = 16

в) $\sigma$ (std(X)) = 4

##### 5. На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от
математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

```python
sigma = (190 - 178) / math.sqrt(25)

print('Количество сигм отклонения роста человека от математического ожидания роста в популяции = {:.4}'.format(sigma))
```
